# Outliers

```python
#find the outliers
outliers=[]
def detect_outliers(df):
    threshold=3
    m=np.mean(df)
    s=np.std(df)
    
    for i in df:
        z_score=(i-m)/s
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_datapoints=detect_outliers(df["Salary"])
```
