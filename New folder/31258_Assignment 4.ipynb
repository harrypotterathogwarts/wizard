# Import libraries

```python
#import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

# Data Visualization

```python
#distribution of price of tickets within 100 bins
sns.histplot(data=df['fare'], bins=100 )
```

```python
#distribution of price of tickets within 10 bins
sns.histplot(data=df['fare'], bins=10 )
```

```python
#display the survival rate of men vs women
sns.catplot(x ="sex", hue ="survived",
kind ="count", data = df)
```

```python
#display how the price of the ticket for each passenger is distributed
sns.distplot(df['fare'])
```

```python
#display how the price of the ticket for each passenger is distributed (without the kernel density line)
sns.distplot(df['fare'], kde=False)
```

```python
##display how the price of the ticket for each passenger is distributed (without the kernel density line and change of bins)
sns.distplot(df['fare'], kde=False, bins=10)
```

```python
#display the mutual distribution between age and fare(usinf scatterplot)
sns.jointplot(x='age', y='fare', data=df)
```

```python
#display the mutual distribution between age and fare(usinf hexagonal plot)
sns.jointplot(x='age', y='fare', data=df, kind='hex')
```

```python
#removing null values
df = df.dropna()
```

```python
#display all the possible combination of numeric and Boolean columns
sns.pairplot(df)
```

```python
#display all the possible combination of numeric and Boolean columns (to plot gender information)
sns.pairplot(df, hue='sex')
```

```python
#display the instances for the fares
sns.rugplot(df['fare'])
```

```python
#display mean value of the age of the male and female passengers
sns.barplot(x='sex', y='age', data=df)
```

```python
##display standard deviation of the age of the male and female passengers
sns.barplot(x='sex', y='age', data=df, estimator=np.std)
```

```python
#display the count of the number of males and women passenger
sns.countplot(x='sex', data=df)
```

```python
#display the distribution for the age with respect to each gender
sns.boxplot(x='sex', y='age', data=df)
```

```python
#display the distribution for the age with respect to each gender along with the information about whether or not they survived
sns.boxplot(x='sex', y='age', data=df, hue="survived")
```

```python
#display the distribution for the age with respect to each gender
sns.violinplot(x='sex', y='age', data=df)
```

```python
#display the distribution for the age with respect to each gender along with the information about whether or not they survived
sns.violinplot(x='sex', y='age', data=df, hue='survived')
```

```python
#display the distribution for the age with respect to each gender
sns.scatterplot(x='sex', y='age', data=df)
```

```python
#display the distribution for the age with respect to each gender along with the information about whether or not they survived
sns.scatterplot(x='sex', y='age', data=df, hue='survived')
```

```python
#create boxplots in each subplot
fig, axes = plt.subplots(2, 2)
sns.boxplot(data=df, x='survived', y='age', ax=axes[0,0])
sns.boxplot(data=df, x='survived', y='fare', ax=axes[0,1])
sns.boxplot(data=df, x='survived', y='parch', ax=axes[1,0])
sns.boxplot(data=df, x='survived', y='pclass', ax=axes[1,1])
```
